O avanço da tecnologia na área da saúde, especialmente com a análise de imagens biomédicas, tem desempenhado um papel crucial no diagnóstico e tratamento de diversas doenças. A análise de imagens de tomografias computadorizadas (CT) e ressonâncias magnéticas (MRI) permite detectar anomalias estruturais em órgãos e tecidos, oferecendo aos médicos informações valiosas para decisões clínicas.

Este projeto concentra-se na aplicação de técnicas de álgebra linear na análise de imagens biomédicas. Especificamente, exploraremos métodos numéricos e operações matriciais para manipular e processar imagens de ressonâncias magnéticas. Serão aplicados conceitos como decomposição de matrizes, autovalores e autovetores, e técnicas de redução de dimensionalidade, como a Decomposição de Valores Singulares (SVD), para destacar estruturas importantes nas imagens. O objetivo final é identificar possíveis alterações cerebrais associadas à doença de Alzheimer, além de realizar segmentações em imagens médicas.

A partir disso, antes de começar a realizar as aplicações, vamos entender como a doença funciona: O Alzheimer é uma doença neurodegenerativa progressiva que afeta principalmente pessoas idosas, resultando em declínio cognitivo, perda de memória, alterações no comportamento e, eventualmente, a incapacidade de realizar atividades cotidianas. A doença é caracterizada por mudanças no cérebro, como a formação de placas de beta-amiloide e emaranhados de proteína tau, que prejudicam a comunicação entre os neurônios e levam à morte dessas células cerebrais.

Fases do Alzheimer:

  * **Sem demência (Normal):**
  Neste estágio, as pessoas não apresentam sinais evidentes de Alzheimer. Elas podem experimentar pequenas mudanças de memória, mas isso é considerado dentro da faixa de envelhecimento normal.
Estruturalmente, o cérebro pode já começar a mostrar algumas das mudanças características do Alzheimer (como pequenas acumulações de placas beta-amiloides), mas as funções cognitivas e comportamentais permanecem estáveis.
  * **Demência leve:**
Neste estágio, os sintomas começam a se tornar mais perceptíveis tanto para a pessoa afetada quanto para familiares. Estruturalmente, as regiões do cérebro responsáveis pela memória, como o hipocampo, começam a mostrar sinais de encolhimento.
  * **Demência moderada:**
Os sintomas tornam-se mais severos e afetam a vida cotidiana. As capacidades cognitivas se deterioram significativamente. Estruturalmente, a perda de tecido cerebral é mais generalizada, e áreas importantes para o pensamento, raciocínio e movimento podem começar a ser afetadas.
À medida que a doença progride, o comprometimento cognitivo e funcional continua a se agravar até o estágio final, em que o paciente perde a capacidade de interagir com o ambiente e depende totalmente de cuidados.

Para resolver esse problema, utilizaremos os seguintes métodos numéricos:

1. **Transformações Lineares em Imagens:**

  * **Operações matriciais:** As imagens biomédicas podem ser vistas como grandes matrizes, onde cada elemento representa a intensidade do pixel. Utilizaremos operações matriciais, como a multiplicação por uma matriz de transformação, para ajustar e filtrar a imagem.

2. **Decomposição de Valores Singulares (SVD):**

  * **Aplicação em redução de ruído:** A SVD será aplicada para decompor a imagem em componentes fundamentais, separando o ruído das informações mais importantes da imagem. Esse método é útil para destacar estruturas cerebrais relevantes que podem estar associadas à doença de Alzheimer.
  * **Autovalores e Autovetores:** Serão usados para identificar as direções principais de variação nas imagens, o que ajuda a localizar regiões afetadas por doenças neurodegenerativas.

3. **Segmentação de Imagens:**

  * **Filtros e convoluções:** Aplicaremos operações de filtragem baseadas em álgebra linear para segmentar regiões específicas nas imagens ressonância, destacando estruturas do cérebro.

  * **Detecção de Bordas com Filtro Sobel:**
O filtro Sobel é um operador de convolução usado para detectar bordas nas imagens. Ele calcula as derivadas das intensidades dos pixels nas direções horizontal e vertical, destacando mudanças abruptas de intensidade que geralmente correspondem às bordas das estruturas.
Na análise de imagens cerebrais, o filtro Sobel é útil para identificar fronteiras entre diferentes regiões do cérebro, como os ventrículos e o córtex, o que é crucial para destacar áreas potencialmente afetadas por atrofia no Alzheimer.
4. **Comparação das Texturas:**
  * **Média de Intensidades:**
A média de intensidades de uma imagem é usada para comparar a textura de diferentes regiões cerebrais. Regiões afetadas por doenças neurodegenerativas, como o Alzheimer, podem apresentar alterações na densidade do tecido cerebral, o que se reflete em mudanças na intensidade média dos pixels.
Comparar essas médias entre diferentes áreas ou entre exames de diferentes pacientes pode ajudar a identificar anomalias sutis.
 * **Comparação da Diferença de Intensidade entre Imagens:**
Ao comparar imagens de ressonância magnética de diferentes períodos ou de diferentes pacientes, pode-se calcular a diferença de intensidade entre as imagens para localizar alterações estruturais ao longo do tempo.
Por exemplo, comparações de imagens de ressonância ao longo do tempo podem revelar a progressiva perda de massa cerebral característica de estágios avançados do Alzheimer.
  * **Comparação dos Tamanhos dos Ventrículos:**
O tamanho dos ventrículos cerebrais pode ser um indicativo de atrofia cerebral. Na doença de Alzheimer, ocorre frequentemente a dilatação dos ventrículos à medida que o tecido cerebral ao redor se deteriora.
A segmentação dos ventrículos e a medição de seu tamanho, com base em operações matriciais, ajudam a quantificar essa dilatação e podem ser usadas para monitorar a progressão da doença.
 * **Análise de Similaridade Estrutural (SSIM):**
O SSIM (Structural Similarity Index) é uma métrica que compara a similaridade estrutural entre duas imagens, levando em consideração aspectos como luminância, contraste e estrutura. Ao invés de apenas comparar a diferença de intensidades pixel a pixel, o SSIM avalia a qualidade visual e a consistência estrutural entre duas imagens.
Na análise de ressonâncias magnéticas cerebrais, o SSIM pode ser usado para comparar a imagem de um cérebro saudável com a de um cérebro afetado pelo Alzheimer, ajudando a identificar e destacar áreas com diferenças estruturais significativas.

Esses métodos são apropriados para lidar com grandes volumes de dados, como os encontrados em imagens médicas, e permitem isolar características importantes, facilitando o diagnóstico.

##**Preparação do Ambiente e Carregamento das Imagens**


Primeiro, instalamos e importamos as bibliotecas necessárias:

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import filters
from scipy.linalg import svd
import imageio
```

Carregamos quatro imagens de Ressonâncias Magnéticas de cérebros distintos para análise:

```python
# Carregando as imagens e verificando se o carregamento foi bem-sucedido

def load_image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Erro ao carregar a imagem: {path}")
    return img

img_dementia_moderate = load_image('/content/moderate_57.jpg')
img_dementia_mild = load_image('/content/mild_6.jpg')
img_dementia_very_mild = load_image('/content/verymild_6.jpg')
img_normal = load_image('/content/non_6.jpg')

# Função para visualizar as imagens

def plot_images(img1, img2, img3, img4, title1, title2, title3, title4):
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 4, 1)
    plt.imshow(img1, cmap='gray')
    plt.title(title1)
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(img2, cmap='gray')
    plt.title(title2)
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(img3, cmap='gray')
    plt.title(title3)
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(img4, cmap='gray')
    plt.title(title4)
    plt.axis('off')

    plt.show()

# Visualizar as imagens

plot_images(img_normal, img_dementia_very_mild, img_dementia_mild, img_dementia_moderate,
            "Cérebro Normal", "Demência Muito Leve", "Demência Leve", "Demência Moderada")
```

##**Análise SVD**

Agora iremos fazer a **Decomposição de Valores Singulares (SVD)** das imagens de ressonância magnética dos cérebros.

A análise SVD é uma técnica fundamental de álgebra linear que decompõe uma matriz (no caso, uma matriz de pixels da imagem) em três componentes principais: `𝑈`, `𝑆`, e `𝑉^𝑇`. Essa colocação é muito útil para reduzir a dimensionalidade e capturar as principais características estruturais da imagem.
Isso nos permitirá visualizar como a decomposição em valores singulares varia entre o cérebro normal e os cérebros com demência em diferentes estágios.

Aqui, a imagem `img` (uma matriz de pixels) é decomposta em três componentes usando a função `svd()`:

* **U:** Uma matriz ortogonal que representa as "direções" principais da imagem no espaço das linhas.
* **S:** Um vetor com os valores singulares da imagem, que indicam a importância de cada componente na proteção da imagem.
* **Vt:** A transposta da matriz ortogonal que descreve as "direções" principais no espaço das colunas.

```python
# Função para análise SVD:
# Essa função executa a configuração SVD para a imagem de entrada e plota os valores singulares.

def svd_analysis(img, title):
    U, S, Vt = svd(img, full_matrices=False)
    plt.figure(figsize=(8, 4))
    plt.plot(S, 'bo-')
    plt.title(f'Valores Singulares - {title}')
    plt.xlabel('Index')
    plt.ylabel('Singular Value')
    plt.grid(True)
    plt.show()
    return S

# Analisar SVD para todas as imagens
svd_normal = svd_analysis(img_normal, "Cérebro Normal")
svd_dementia_very_mild = svd_analysis(img_dementia_very_mild, "Demência Muito Leve")
svd_dementia_mild = svd_analysis(img_dementia_mild, "Demência Leve")
svd_dementia_moderate = svd_analysis(img_dementia_moderate, "Demência Moderada")

# Criar a figura e os subplots
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
# Plotar os gráficos
axes[0].plot(svd_normal, 'bo-')
axes[0].set_title('Cérebro Normal')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Singular Value')
axes[0].grid(True)

axes[1].plot(svd_dementia_very_mild, 'bo-')
axes[1].set_title('Demência Muito Leve')
axes[1].set_xlabel('Index')
axes[1].grid(True)

axes[2].plot(svd_dementia_mild, 'bo-')
axes[2].set_title('Demência Leve')
axes[2].set_xlabel('Index')
axes[2].grid(True)

axes[3].plot(svd_dementia_moderate, 'bo-')
axes[3].set_title('Demência Moderada')
axes[3].set_xlabel('Index')
axes[3].grid(True)

# Ajustar o layout
plt.tight_layout()
plt.show()
```

Criamos um gráfico dos valores singulares
`𝑆` para a imagem correspondente. Isso ajuda a visualizar como os valores singulares decrescem.
* **Eixo X :** O índice do valor singular.
* **Eixo Y :** O valor singular em si.

Esses valores singulares ajudam a descrever como as características estruturais do cérebro mudam em cada estágio de demência.


##**Detectar bordas usando o filtro Sobel**
Aqui, executamos a função `sobel_edges` para cada uma das quatro imagens e visualizamos as bordas detectadas. A função retorna uma versão de cada imagem em que as bordas foram detectadas, destacando as áreas de transição na estrutura cerebral.


Utilizamos a detecção de bordas através do filtro de Sobel para encontrar e visualizar as bordas (mudanças abruptas de intensidade) nas imagens de ressonância magnética dos cérebros. A detecção de bordas é uma técnica essencial no processamento de imagens, especialmente para destacar estruturas e contornos importantes, o que é muito útil na análise de imagens biomédicas.

```python
# Comparação de bordas e gradientes usando filtro de Sobel

def sobel_edges(img, title):
    edges = filters.sobel(img)
    plt.figure(figsize=(6, 6))
    plt.imshow(edges, cmap='gray')
    plt.title(f'Bordas Detetadas com Sobel - {title}')
    plt.axis('off')
    plt.show()
    return edges

# Aplicar detecção de bordas
edges_normal = sobel_edges(img_normal, "Cérebro Normal")
edges_dementia_very_mild = sobel_edges(img_dementia_very_mild, "Demência Muito Leve")
edges_dementia_mild = sobel_edges(img_dementia_mild, "Demência Leve")
edges_dementia_moderate = sobel_edges(img_dementia_moderate, "Demência Moderada")
```

O filtro de Sobel é uma técnica clássica de detecção de bordas, que calcula uma aproximação dos gradientes de intensidade da imagem nas específicas horizontais e verticais.
* O gradiente mede como a intensidade dos pixels muda de um ponto para o outro na imagem. Onde houver uma mudança brusca de intensidade (uma borda), o gradiente será grande.
* O método `filters.sobel()` aplica esse filtro na imagem `img`, retornando uma nova imagem `edges` que contém a intensidade das bordas detectadas.
* Na prática, as bordas destacam regiões da imagem onde há transições rápidas de cor ou intensidade (como a transição entre diferentes tecidos no cérebro).

Através dos resultados obtidos, podemos observar que:

* **Cérebro Normal :** As bordas detectadas realçam as estruturas normais, como a distinção entre as substâncias brancas e cinzenta, e a anatomia regular dos giros e sulcos.

* **Cérebro com Demência Muito Leve, Leve e Moderada :** As bordas detectadas para essas imagens nos mostraram uma interrupção ou alteração estrutural, como atrofia cortical (diminuição da espessura da substância cinzenta) e a dilatação dos ventrículos (cavidades no cérebro), que são sinais típicos da substância doença de Alzheimer.

##**Comparação das texturas: média de intensidades**


```python
# Comparação das texturas: média de intensidades
mean_intensity_normal = np.mean(img_normal)
mean_intensity_very_mild = np.mean(img_dementia_very_mild)
mean_intensity_mild = np.mean(img_dementia_mild)
mean_intensity_moderate = np.mean(img_dementia_moderate)

print(f"Média de Intensidade - Cérebro Normal: {mean_intensity_normal}")
print(f"Média de Intensidade - Demência Muito Leve: {mean_intensity_very_mild}")
print(f"Média de Intensidade - Demência Leve: {mean_intensity_mild}")
print(f"Média de Intensidade - Demência Moderada: {mean_intensity_moderate}")
```

##**Comparação da diferença de intensidade entre as imagens**

Agora, mostramos um mapa de calor das diferenças de intensidade. Regiões com cores mais quentes indicam áreas onde as intensidades diferem mais entre as imagens.

```python
# Comparar a diferença de intensidade entre as imagens, comparando com a normal
diff_intensity_very_mild = np.abs(img_dementia_very_mild.astype(np.float32) - img_normal.astype(np.float32))
diff_intensity_mild = np.abs(img_dementia_mild.astype(np.float32) - img_normal.astype(np.float32))
diff_intensity_moderate = np.abs(img_dementia_moderate.astype(np.float32) - img_normal.astype(np.float32))

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(img_normal, cmap='gray')
plt.title('Diferença de Intensidade - Normal')
plt.colorbar()
plt.axis('off')

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(diff_intensity_very_mild, cmap='hot')
plt.title('Diferença de Intensidade - Muito Leve')
plt.colorbar()
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(diff_intensity_mild, cmap='hot')
plt.title('Diferença de Intensidade - Leve')
plt.colorbar()
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(diff_intensity_moderate, cmap='hot')
plt.title('Diferença de Intensidade - Moderada')
plt.colorbar()
plt.axis('off')

plt.show()
```

##**Comparação dos tamanhos dos ventrículos:**

Agora, comparamos o número de pixels mais escuros entre o cérebro normal e os cérebros com demência, permitindo ver como os ventrículos aumentam com a progressão da doença.

```python
# Comparar a variação do tamanho dos ventrículos
ventricles_normal = np.sum(img_normal < 50)  # Pixels mais escuros (ventrículos)
ventricles_very_mild = np.sum(img_dementia_very_mild < 50)
ventricles_mild = np.sum(img_dementia_mild < 50)
ventricles_moderate = np.sum(img_dementia_moderate < 50)

print(f"Tamanho dos ventrículos - Cérebro Normal: {ventricles_normal} pixels")
print(f"Tamanho dos ventrículos - Demência Muito Leve: {ventricles_very_mild} pixels")
print(f"Tamanho dos ventrículos - Demência Leve: {ventricles_mild} pixels")
print(f"Tamanho dos ventrículos - Demência Moderada: {ventricles_moderate} pixels")
```

##**Análise de Similaridade Estrutural (SSIM)**

Aqui, fazemos a comparação entre diferentes estágios: Cada chamada compara duas imagens em termos de similaridade estrutural:
* Normal vs Demência Muito Leve.
* Normal vs Demência Leve.
* Normal vs Demência Moderada.
* Demência Muito Leve vs Demência Leve.
* Demência Leve vs Demência Moderada.

Esses resultados mostram como a estrutura do cérebro muda conforme a demência progride. O valor SSIM nos permite quantificar essa mudança, enquanto o mapa visual nos ajuda a identificar as regiões com maior ou menor mudança.

```python
#Análise de Similaridade Estrutural (SSIM):

from skimage.metrics import structural_similarity as ssim

def ssim_analysis(img1, img2, title1, title2):
    ssim_index, ssim_map = ssim(img1, img2, full=True)
    print(f"SSIM entre {title1} e {title2}: {ssim_index:.4f}")
    plt.figure(figsize=(6, 6))
    plt.imshow(ssim_map, cmap='viridis')
    plt.title(f"Mapa SSIM entre {title1} e {title2}")
    plt.colorbar()
    plt.axis('off')
    plt.show()
    return ssim_index


ssim_analysis(img_normal, img_dementia_very_mild, "Cérebro Normal", "Demência Muito Leve")
ssim_analysis(img_normal, img_dementia_mild, "Cérebro Normal", "Demência Leve")
ssim_analysis(img_normal, img_dementia_moderate, "Cérebro Normal", "Demência Moderada")
ssim_analysis(img_dementia_very_mild, img_dementia_mild, "Demência Muito Leve", "Demência Leve")
ssim_analysis(img_dementia_mild, img_dementia_moderate, "Demência Leve", "Demência Moderada")
```
